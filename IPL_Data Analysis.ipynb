# EDA (Exploratory Data Analysis)

```python
df.head(2)
```

```python
top_scores = df.sort_values(by='Runs',ascending=False).head(10)
top_scores
```

# Top 10 Run Scorers

```python
# top_scores.plot(x='Player Name',y='Runs',kind='bar',color='orange',title='Top 10 Run Scores');
plt.figure(figsize=(10, 6))

sns.barplot(x='Player Name',y='Runs',data = top_scores ,palette='Set1')
plt.title('Top 10 Run Scores',fontsize=14)
plt.ylabel('Runs')
plt.xlabel('Player Name')
plt.xticks(rotation=90)
plt.show()
```

# Top Strike Rates

```python
df_filtered = df[df['SR']>100]
df_filtered
```

```python
Top_Sr = df_filtered.sort_values(by='SR',ascending=False).head(10)
Top_Sr
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='Player Name',y='SR',data = Top_Sr,palette='Set1')
plt.title('Top 10 Strike Rate',fontsize=14)
plt.ylabel('Strike Rate')
plt.xlabel('Player Name')
plt.xticks(rotation=90)
plt.show()

```

# Highest Batting Score

```python
Top_Hs = df_filtered.sort_values(by='HS',ascending=False).head(10)
Top_Hs
```

```python
plt.figure(figsize=(10,6))

plt.title('Top 10 Highest Score')
sns.barplot(x='Player Name',y='HS',data=Top_Hs,palette='Set1')

plt.xlabel('Player Name')
plt.ylabel('Highest Score')
plt.xticks(rotation=90)
plt.gca().invert_yaxis()
plt.show()
```

# Highest Batting Averages Score

```python
Top_AVG = df_filtered.sort_values(by='AVG',ascending=False).head(10)
Top_AVG
```

```python
plt.figure(figsize=(10,7),dpi=500)

sns.barplot(x='Player Name',y='AVG',data=Top_AVG,palette='Set1')
plt.title('TOP 10 Average Rate',fontsize=14)
plt.xlabel('Player_Name')
plt.ylabel('Average')
plt.xticks(rotation=90)
plt.gca().invert_yaxis()
plt.show()
```

# Distribution of Batting Average (Histogram)

```python
sns.histplot(df['AVG'],bins=20,kde=True,color='skyblue')
plt.title('Distribution of Batting Average')
plt.xlabel('Average')
plt.ylabel('Number of plyers')
plt.show()
```

#  Scatter Plot: AVG vs SR

```python
sns.scatterplot(x='AVG', y='SR', hue='Team', data=df, palette='tab10')
plt.title('Batting Average vs Strike Rate')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.show()
```

# Boundary Hitters (Most 4s and 6s)

#  Best Teams by Total Runs

```python
df.head(2)
```

```python
Top_Team = df.groupby(['Team'])['Runs'].sum().sort_values(ascending=False).head(10).reset_index()
Top_Team
```

```python
plt.figure(figsize=(10,7),dpi=500)

sns.barplot(x='Team',y='Runs',data=Top_Team,palette='Set1')
plt.title('Top 10 teams Highest Score',fontsize=14)
plt.xlabel('Teams')
plt.ylabel('Runs')
plt.xticks(rotation=90)
plt.show()
```

# Team-wise Average Strike Rate


```python
df.head(2)
```

```python
Team_sr = df.groupby(['Team'])['SR'].mean().sort_values(ascending=False).reset_index()
Team_sr
```

```python
plt.figure(figsize=(10,7),dpi=500)

sns.barplot(x='Team',y='SR',data=Team_sr,palette='Set1')
plt.title('Team wise Average Strike',fontsize=14)
plt.xlabel('Team')
plt.ylabel('Strike Rate')
plt.xticks(rotation=90)
plt.show()
```

# Consistency Score

```python
df['consistency'] = df['50s'] +df['100s'] / df['Inn']
df['consistency']
```

```python
df.head(2)
```

```python
con = df.sort_values(by='consistency',ascending=False).head(10)
con
```

```python
plt.figure(figsize=(10,7),dpi=500)

sns.barplot(x='Player Name',y='consistency',data=con,palette='Set1')
plt.title('Consistency Score',fontsize=14)
plt.xlabel('Player Name')
plt.ylabel('Consistency')
plt.xticks(rotation=90)
plt.show()
```

# Team Strength Index (Batting)

```python
team_strength = df.groupby(['Team']).agg({'Runs':'sum','SR':'mean','4s':'sum','6s':'sum'}).reset_index()
team_strength
```

```python
plt.figure(figsize=(10,7),dpi=500)

sns.barplot(x='Team',y='Runs',data=team_strength,palette='Set1')
plt.title('Team Strength Index', fontsize=14)
plt.xlabel('Team Name')
plt.ylabel('Runs')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,7),dpi=500)

sns.barplot(x='Team',y='SR',data=team_strength,palette='Set1')
plt.title('Team Strength Index', fontsize=14)
plt.xlabel('Team Name')
plt.ylabel('Runs')
plt.xticks(rotation=90)
plt.show()
```
